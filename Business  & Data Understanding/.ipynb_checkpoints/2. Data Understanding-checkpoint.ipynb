# Data requirements 

The first stage of the data requirements is to make sure the data can be used to solve the business problem identified in the Business Understanding. 
There are some issues / requirements to be considered for the data. 

- The first requirement is to set up a concise and clear description of the problem. As mentioned in step Business Understanding, we seek to identify patterns in our data from shares to build a recommendation algorithm for the decision-making of share purchases.
- The second requirement would be to identify the relevant data for the problem description. The opening and closing price of the share, the daily high and low of the share and the volume will be relevant for our purpose. Other fundamental data to be considered is EBIT, revenue, net profit, dividends, etc.
- The third requirement is that selected variables for the relevant data should be independent of each other. Variable independence means that the variables do not contain overlapping information. A careful selection of independent variables can make it easier for data mining algorithms to quickly discover useful knowledge patterns.
- In order to be able to make a forecast, we need the data of the individual shares on a daily basis.

# Data exploration

We identified some data issues:

- Fundamental data is not available on a daily basis, only on a quarterly basis.
- The above quarterly fundamentals are not complete. For instance, the AAPL ticker has a gap between 2014 and 2018, there are is no fundamental data available.
- As defined in the data requirements, a selection of independent variables can make it easier for data mining algorithms to quickly discover useful knowledge patterns. For our purpose this is quite difficult to achive, given that revenue, EBIT, net profit and dividends all depend on each other.

In order to explore the data we check if there are any missing values or outliers in the data. If there are any missing values or outliers we need to make sure there are valid explanations for them.

```python
# Visual inspection for outliers with boxplots

# get numeric columns to check
cols = df.select_dtypes(include=[np.number]).columns
fig, axes = plt.subplots(nrows=len(cols), ncols=1)
# draw boxplot for each col ... need index to address axes
for i in range(len(cols)):
    sns.boxplot(x=cols[i], orient='h', data=df, ax=axes[i])

# improve spacing and show
fig.tight_layout()
plt.show()
```

There are some outliers in stock value due to the opening and closing value being very high in the last stages of the data, there are also some outliers in trading volume, which can occur after certain events, such as an analyst reviews or the result of a quarter being published, or simply if the stock is gaining momentum and the trading increases.

# Verification of data quality

In order to verify the data quality we compared the data with other data sources - such as onvista.de, Google Finance and Yahoo Finance - and found that the data for APPL, ASNA, CAT, EXPE, GMI, IBM, IMO, MCO, NFG, PLD is partly correct:

- Open, close, volume low and high data is correct and also complete on a daily basis.
- Fundamental data, such as EBIT, revenue, net profit and dividends is correct, however, it is not complete on a daily basis.

Due to the fundamental data not being complete it will be left out from now on and we will continue using just open, close, volume low and high data. Furthermore there is no variable independence between some of the fundamental data, for instance the EBIT also depends on the revenue.
